# Subgroup Column

# 1. Total Rape Victims

```python
tr = df[df['Subgroup'] == 'Total Rape Victims']
tr
```

```python
cols=['Victims_Upto_10_Yrs','Victims_Between_10-14_Yrs','Victims_Between_14-18_Yrs',
     'Victims_Between_18-30_Yrs','Victims_Between_30-50_Yrs','Victims_Above_50_Yrs']
```

```python
for i in cols:
    print(f" Total rape victims in India from 2001 to 2010 where {i} was {tr[i].sum()}.")
    print('\n')
```

```python
tr['Rape_Cases_Reported'].sum()
```

### Total rape cases reported in India from 2001 to 2010 was 190008.

```python
tr['Victims_of_Rape_Total'].sum()
```

### Total number of rape victims in India from 2001 to 2010 was 190154.

# Yearly Basis.

```python
yearly_basis=tr.groupby('Year')['Rape_Cases_Reported','Victims_of_Rape_Total'].sum().reset_index()
```

```python
yb=pd.DataFrame(yearly_basis)
```

```python
yb
```

```python
# on average per-day rape cases reported in India froom 2001 to 2010.
yb['Rape_Cases_Reported']/365 
```

```python
# on average per-day rape cases count in India froom 2001 to 2010.
yb['Victims_of_Rape_Total']/365
```

```python
plt.figure(figsize=(10,5))
plt.bar(yb['Year'],yb['Rape_Cases_Reported'])
plt.show()
```

### Maximum number of rape cases were reported in the year 2010

```python
plt.figure(figsize=(10,5))
plt.bar(yb['Year'],yb['Victims_of_Rape_Total'])
plt.show()
```

### Rape cases have been steadily rising in India from 2001 to 2010.

```python
yearly_basis1=tr.groupby('Year')[cols].sum().reset_index()
```

```python
yb1=pd.DataFrame(yearly_basis1)
yb1
```

```python
plt.figure(figsize=(10,5))
plt.bar(yb1['Year'],yb1['Victims_Upto_10_Yrs'])
plt.show()
```

### Maximum number of rape cases where Victims_Upto_10_Yrs was in the year 2009.
