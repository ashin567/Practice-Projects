Removing Outliers

1. Z-Score Method

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
# Threshold = 3
wd = df[(z<3).all(axis=1)]
```

```python
wd.shape
```

```python
df.shape
```

```python
# percentage loss of data
per_loss=((178-168)/178)*100
per_loss
```

2. IQR Method

```python
Q3 = df.quantile(.75)
Q1 = df.quantile(.25)
IQR=Q3-Q1
IQR
```

```python
wd1 = df[~((df < (Q1-1.5*IQR)) | (df > (Q3+1.5*IQR))).any(axis=1)]
```

```python
wd1.shape
```

```python
# percentage loss of data
per_loss=((178-161)/178)*100
per_loss
```

    Since percentage loss of data after removing outliers is maximum in IQR method compared to Z-score method, it is better to choose Z-score method.

```python
x=power_transform(x,method='yeo-johnson')
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x=sc.fit_transform(x)
```

# Model Training And Testing

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# finding the best random_state

acc=0
for i in range(20,800):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    dtc=DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    y_pred=dtc.predict(x_test)
    acc_score=accuracy_score(y_test,y_pred)
    if acc_score>acc:
        acc=acc_score
        best_random_state=i
        
        
print(f"Accuracy score is {acc*100} for random_state {best_random_state}")   
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=best_random_state)
```

```python
# finding the best cv

cv=0
for i in range(2,10):
    cv_score=cross_val_score(dtc,x,y,cv=i).mean()
    if cv_score>cv:
        cv=cv_score
        best_cv=i
        
print(f" Cross_val_score is {cv*100} for cv {best_cv}")    
```

# 1. Decision Tree Classifier

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print('Accuracy score:',acc*100)
cv_score=cross_val_score(dtc,x,y,cv=best_cv).mean()
print('Cross_val_score:',cv_score*100)
print('Difference:',((acc*100)-(cv_score*100)))
print('\n')
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# 2. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
acc=accuracy_score(y_test,y_pred)
print('Accuracy score:',acc*100)
cv_score=cross_val_score(rfc,x,y,cv=best_cv).mean()
print('Cross_val_score:',cv_score*100)
print('Difference:',((acc*100)-(cv_score*100)))
print('\n')
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
